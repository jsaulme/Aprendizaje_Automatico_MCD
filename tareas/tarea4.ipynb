```python
import pandas as pd

data = pd.read_excel("modified_Global_Superstore_2018.xlsx")
data.head(3)
```

```python
variables_numericas = data.select_dtypes(include=[float, int])
variables_numericas.head()
variables_numericas.drop("Index", axis=1, inplace=True)
variables_numericas.head()
```

```python
# Diccionario de abreviaciones
abreviados = {
    'Delivery Time': 'DT',
    'Discount': 'D',
    'Order Priority Numeric': 'OPN',
    'Profit': 'P',
    'Profit Margin': 'PM',
    'Quantity': 'Q',
    'Sales': 'S',
    'Sales per Quantity': 'SQ',
    'Shipping Cost': 'SC',
    'Total Profit': 'TP',
    'Total Sales': 'TS'
}

# Extraer los nombres de las columnas y aplicar las abreviaciones
var_num_columnas = variables_numericas.columns.tolist()
nombres_abreviados = [abreviados[name] for name in var_num_columnas]

# Asignar los nombres abreviados a las columnas
variables_numericas.columns = nombres_abreviados
```

```python
variables_numericas.head(3)
```

Aplica algun metodo de filtro a tus datos mediante el uso de #SelectKBest#

```python
# Aplica metodo de filtro al dataset usando SelectKBest

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression

# Define tus datos 'x' (características) y 'y' (variable objetivo)
x = variables_numericas.drop(columns=['TS'])            # DataFrame con todas las características numéricas, excepto la variable objetivo
y = variables_numericas['TS']                           # Serie que contiene la variable objetivo que deseas predecir

# Define el método de selección de características
selector = SelectKBest(score_func=f_regression, k=4)

# Ajusta el selector a los datos 'x' y 'y'
selector.fit(x, y)

# Transforma los datos utilizando el selector
X_new = selector.transform(x)

# Imprime las características seleccionadas
print("Características seleccionadas:", x.columns[selector.get_support()])
```

# Determina las características mas relevantes segun tu conjunto de datos

Estos son los modelos aplicados y sus resultados al destacar caractersítcicas:

Regresion F (Anova):
Total profit, Sales, Sales Per Quantity;

Correlacion: Total Profit, Sales Per Quantity, Shipin Cost;


Seleccion de caracteristicas exhaustiva: Sales, Sales per Quantity, Total Profit;


Sequential Forward Selection (SFS): 'Sales', 'Sales per Quantity', 'Total Profit

Para analizar por qué estas características fueron seleccionadas como las más relevantes en el dataset, es necesario considerar el contexto del dataset y  los métodos de selección de características utilizados.


1. Total Profit
Es una métrica clave en todo análisis de ventas, dado que refleja el beneficio generado por las transacciones. En un análisis de ventas, maximizar el beneficio es fundamental.
Fue seleccionada por Regresión F (Anova), Correlación, Selección de características exhaustiva y SFS.
Esta métrica captura el resultado neto de las ventas y descuentos, haciendo que sea una variable esencial para el análisis predictivo relacionado con ventas y rentabilidad.
2. Sales (Ventas)
Las ventas son otro indicador crucial de desempeño. No solo reflejan la cantidad de ingresos sino también la demanda de productos.
Seleccionada por Regresión F (Anova), Selección de características exhaustiva y SFS.
Las ventas proporcionan una base para entender el volumen de transacciones y el impacto de las estrategias de marketing y promociones.
3. Sales per Quantity
Esta característica normaliza las ventas por la cantidad de productos vendidos, señalando una medida de eficiencia en términos de ingresos por unidad vendida. Seleccionada por todos los métodos (Regresión F, Correlación, Selección exhaustiva y SFS).
Ofrece una perspectiva más granular sobre la relación entre ventas y cantidad, lo que ayuda a identificar estrategias de precios efectivas y tendencias de compra.
4. Shipping Cost
Los costos de envío afectan tanto a los márgenes de beneficio como a la satisfacción del cliente. Es un factor importante en la gestión de la cadena de suministro y la optimización de costos.
Modelos: Seleccionada por Correlación.
Este costo puede variar significativamente entre pedidos y afecta directamente al costo total y al margen de beneficio.


Exclusión del Resto de Características
1. Delivery Time
Aunque el tiempo de entrega es importante para la satisfacción del cliente, puede no tener una correlación directa tan fuerte con las métricas financieras principales como el beneficio o las ventas.
2. Discount
Los descuentos son estrategias tácticas que impactan temporalmente las ventas y los beneficios. Sin embargo, su impacto puede ser indirecto o moderado cuando se analizan en conjunto con las ventas y los beneficios.
3. Order Priority Numeric
Aunque la prioridad de los pedidos podría influir en la logística y la satisfacción del cliente, su impacto directo en las métricas financieras podría ser menos pronunciado en comparación con las ventas y los beneficios.
4. Quantity
Aunque la cantidad de productos vendidos es relevante, esta métrica está estrechamente relacionada con "Sales" y "Sales per Quantity", lo que puede causar redundancia. Los modelos pueden haber identificado "Sales per Quantity" como una métrica más eficiente para representar la misma información.
