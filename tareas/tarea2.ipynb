```python
# Carga usando el nombre correcto del archivo subido
data = pd.read_excel("Global Superstore 2018.xlsx")

# Mostrar las primeras filas
data.head()
```

## Asignar una columna como indice

```python
# Agregar una columna como indice

data["Index"] = np.arange(len(data))
data = data.set_index("Index")
```

## Crear, modificar o eliminar columnas.

```python
# Transformar el tipo de columna a datetime

data["Order Date"] = pd.to_datetime(data["Order Date"])
data["Ship Date"] = pd.to_datetime(data["Ship Date"])
```

```python
data.sample(5, replace=False)
```

## Crear columnas

```python
# Asegurar que 'Order Date' y 'Ship Date' esten en formato datetime
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])

# Crea la columna 'Delivery Time'. El resultado estará en días
data['Delivery Time'] = (data['Ship Date'] - data['Order Date']).dt.days
```

```python
data[data['Delivery Time']>5]
```

```python
data.columns[2:4]
```

```python
# Crear una lista de las columnas
cols = list(data.columns)

# Eliminar 'Delivery Time' de la lista de columnas
cols.remove('Delivery Time')

# Reorganizar las columnas insertando 'Delivery Time'
data = data[cols[:4] + ['Delivery Time'] + cols[4:]]
data.head()
```

```python
# Calcular ventas por cantidad
data['Sales per Quantity'] = data['Sales'] / data['Quantity']
data["Sales per Quantity"]
```

```python
# Calcular margen de beneficio
data['Profit Margin'] = data['Profit'] / data['Sales']
data['Profit Margin']
```

```python
# Calcular el Customer Lifetime Value (CLV) sumando todas las ventas y ganancias por cliente
clv = data.groupby('Customer ID').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
clv.columns = ['Customer ID', 'Total Sales', 'Total Profit']
data = data.merge(clv, on='Customer ID', how='left')
```

```python
# Mostrar las primeras filas del grupo 'clv'
print("Customer Lifetime Value (CLV) grupo:")
print(clv.head())
```

```python
# Mostrar las primeras filas del dataframe modificado
data.head()
```

```python
# Calcular los cuartiles para la columna 'Shipping Cost'
Q1 = data['Shipping Cost'].quantile(0.25)
Q2 = data['Shipping Cost'].quantile(0.50)  # Esto es también conocido como la mediana
Q3 = data['Shipping Cost'].quantile(0.75)
min_value = data['Shipping Cost'].min()
max_value = data['Shipping Cost'].max()

# Definir los límites (bins) usando el mínimo, los cuartiles y el máximo
bins = [min_value, Q1, Q2, Q3, max_value]
labels = ['Low', 'Medium', 'High', 'Very High']

# Categorizar los costos de envío en rangos
data['Shipping Category'] = pd.cut(data['Shipping Cost'], bins=bins, labels=labels, include_lowest=True)
data.head()
```

## Columnas a Modificar

```python
# Mapeo de las categorías a numeros
priority_mapping = {'Low': 1, 'Medium': 2, 'High': 3, 'Critical': 4}

# Asignar los valores numéricos a la columna 'Order Priority'
data['Order Priority Numeric'] = data['Order Priority'].map(priority_mapping)

# Verificar los registros para confirmar cambios
print(data[['Order Priority', 'Order Priority Numeric']].head())
```

## Filtrar Registros

```python
# Filtrar registros con ventas mayores a 1000
high_sales = data[data['Sales'] > 1000]
print("Registros con ventas mayores a 1000:")
high_sales.head()
```

```python
# Filtrar registros de la categoría 'Technology' con margen de beneficio mayor al 20%
high_margin_tech = data[(data['Category'] == 'Technology') & (data['Profit Margin'] > 0.2)]
print("Registros de la categoría 'Technology' con margen de beneficio mayor al 20%:")
high_margin_tech.head()
```

```python
# Filtrar registros de Estados Unidos en la categoría 'Furniture'
us_furniture = data[(data['Country'] == 'United States') & (data['Category'] == 'Furniture')]
print("Registros de Estados Unidos en la categoría 'Furniture':")
us_furniture.head()
```

## Grupos de Datos

```python
# Agrupar por regin y sumar las ventas totales en cada regin
group_by_region_sales = data.groupby('Region')['Sales'].sum()

# Ordenar los resultados descendente para el total de ventas
group_by_region_sales = group_by_region_sales.sort_values(ascending=False)

# Seleccionar las regiones con mayores ventas
top_regions_by_sales = group_by_region_sales.head(3)

# Mostrar el grupo con las regiones con mayor número de ventas
print(top_regions_by_sales)
```
